**1. Create a DataFrame birds from this dictionary data which has the index labels.**

```python
data = {'birds': ['Cranes', 'Cranes', 'plovers', 'spoonbills', 'spoonbills', 'Cranes', 'plovers', 'Cranes', 'spoonbills', 'spoonbills'], 'age': [3.5, 4, 1.5, np.nan, 6, 3, 5.5, np.nan, 8, 4], 'visits': [2, 4, 3, 4, 3, 4, 2, 2, 3, 2], 'priority': ['yes', 'yes', 'no', 'yes', 'no', 'no', 'no', 'yes', 'no', 'no']}

#Always insert the data into a dataframe (for matrix) or series (for a single array)

import pandas as pd
import numpy as np

change_labels = pd.DataFrame(data, index = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j'])

print(change_labels)
```

**11. Calculate the mean age for each different birds in dataframe.**

```python
mea = change_labels.groupby(['birds'])
print(mea['age'].agg(np.mean))
```

**12. Append a new row 'k' to dataframe with your choice of values for each column. Then delete that row to return the original DataFrame.**

```python
df2 = {'birds':'crane', 'age': 20, 'visits':5, 'priority':'n'}

# appending row to original dataframe
c = change_labels.append(df2, ignore_index =True)
print(c)

print(50*'#')

# deleting or dropping the appended row from new dataset with index number
c.drop([10], inplace= True)
print(c)
```

**13. Find the number of each type of birds in dataframe (Counts)**

**14. Sort dataframe (birds) first by the values in the 'age' in decending order, then by the value in the 'visits' column in ascending order.**

```python
#sorting by age in descending order
so = change_labels.sort_values(by = ['age'], ascending= False)
print(so)

print(50*'#')
# in sorting by visits in ascending order
so.sort_values(by=['visits'])
```

**15. Replace the priority column values with'yes' should be 1 and 'no' should be 0**

```python
# using loc to replace values in a column

# note- when we are trying to save edited or replaced dataframe to a new variable it not saving properly

change_labels.loc[(change_labels.priority == 'yes'), 'priority'] = 1

change_labels.loc[(change_labels.priority == 'no'), 'priority'] = 0

change_labels
```

**16. In the 'birds' column, change the 'Cranes' entries to 'trumpeters'.**
